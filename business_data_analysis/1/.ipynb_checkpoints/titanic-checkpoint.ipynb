```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv", sep=",")
df.head()
```

```python
# We are converting 1912 dollars to 2020 dollars (26.6 is the rate)
# https://www.in2013dollars.com/us/inflation/1912
```

```python
df["Fare 2020"] = df["Passenger Fare"] * 26.6
df.head()
```

```python
# Andrews, Mr. Thomas Jr is the creator of Titanic
df.sort_values(by=['Fare 2020', "Name"], ascending=True, na_position='first')
```

```python
df_grouped = df.groupby(["Passenger Class", "Sex"]).mean()
df_grouped.head()
```

```python
# With NaN's
# Or .fillna(0)
df.astype(str).groupby(['Life Boat']).agg('count')
```

```python
# Without NaN's
df.groupby("Life Boat").agg(["count"])
```
